```python
def get_data(aTuple):
    '''
    it's written on lecture note.
    '''
    nums = ()
    words = ()

    for t in aTuple:
        nums = nums + (t[0],)
        if t[1] not in words:
            words = words + (t[1],)

    min_n = min(nums)
    max_n = max(nums)
    u_words = len(words)

    return(min_n, max_n, u_words)

aTuple = ((3, 'dgu'), (10, 'seoul'), (2, 'ise'), (12, 'dgu'))

(min_n, max_n, u_words) = get_data(aTuple)
```
