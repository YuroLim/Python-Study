```python
def mult_iter(a,b):
    result = 0
    while b > 0:
        result += a
        b -= 1
    return result

print(mult_iter(3,4))
```

```python
def fact(n):
    if n ==1:
        return 1
    else:
        return n*fact(n-1)

print(fact(4))
```

```python
def fib(x):
    """
    assumes x an int >= 0
    returns Fibonacci of x
    """
    if x ==0 or x ==1:
        return 1
    else:
        return fib(x-1) + fib(x-2)

print(fib(5))
```
